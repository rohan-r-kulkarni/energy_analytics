```python
## https://hydrofunctions.readthedocs.io/en/master/introduction.html
import hydrofunctions as hf
import pandas as pd
# start_date = "2000-01-01"
start_date = "2000-01-01"
end_date = "2022-12-31"
```

```python
## https://help.waterdata.usgs.gov/code/parameter_cd_query?fmt=rdb&inline=true&group_cd=%
## 00400	Physical	pH, water, unfiltered, field, standard units	Agree					Total		 	pH	std units
## 63680	Physical	Turbidity, water, unfiltered, monochrome near infra-red LED light, 780-900 nm, detection angle 90 +-2.5 degrees, formazin nephelometric units (FNU)	USGS use and no use by EPA					Total		 	Turbidity	FNU
## 00010	Physical	Temperature, water, degrees Celsius	Agree							 	Temperature, water	deg C
```

```python
parameters = ["00400","63680","00010"]
```

```python
california = hf.NWIS(stateCd = 'CA', start_date=start_date, end_date=end_date, parameterCd=parameters)
```

```python
california_df = california.df()
california_df.head()
```

```python
california_df = california_df.reset_index()
```

```python
california_df.head()
```

```python
california_df.columns
```

```python
california_df.columns
```

```python
# print(california_df.columns.str.contains('00400'))
```

```python
# get turbidity values
turbidity_columns = [col for col in california_df.columns if ':63680' in col and not col.endswith('_qualifiers')]
turbidity = california_df[turbidity_columns].mean(axis=1)

# get temperature values
temperature_columns = [col for col in california_df.columns if ':00010' in col and not col.endswith('_qualifiers')]
temperature = california_df[temperature_columns].mean(axis=1)

# get pH values
ph_columns = [col for col in california_df.columns if ':00400' in col and not col.endswith('_qualifiers')]
ph = california_df[ph_columns].mean(axis=1)

# water quality for CA
water_quality_CA = pd.DataFrame({'datetimeUTC': california_df['datetimeUTC'], 
                         'turbidity value': turbidity,
                         'temperature value': temperature,
                         'pH value': ph})
```

```python
water_quality_CA
```

```python
water_quality_CA['datetimeUTC'] = pd.to_datetime(water_quality_CA['datetimeUTC'])
water_quality_CA = water_quality_CA.set_index('datetimeUTC')
df_monthly = water_quality_CA.resample('M').mean()
```
